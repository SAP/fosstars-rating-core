```python
import commons


# this is a list of features which are used in the Community Commitment Score
features = [
    'How well security testing is done for an open-source project',
    'How well vulnerabilities are patched',
    'How fast vulnerabilities are patched',
    'How well open-source community is aware about security',
    'Open-source project activity score',
    'Open-source project popularity score',
    'How well open-source community commits to support an open-source project'
]


class TestVector(commons.BaseTestVector):
        
    def __init__(self, alias = ''):
        super().__init__(features, alias)

    def security_testing_score(self, value):
        return self.set('How well security testing is done for an open-source project', value)
    
    def unpatched_vulnerabilities_score(self, value):
        return self.set('How well vulnerabilities are patched', value)
    
    def vulnerability_lifetime_score(self, value):
        return self.set('How fast vulnerabilities are patched', value)
    
    def security_awareness_score(self, value):
        return self.set('How well open-source community is aware about security', value)
    
    def activity_score(self, value):
        return self.set('Open-source project activity score', value)
    
    def popularity_score(self, value):
        return self.set('Open-source project popularity score', value)
    
    def community_commitment_score(self, value):
        return self.set('How well open-source community commits to support an open-source project', value)
    

test_vector_list = commons.TestVectorList(features)
```
