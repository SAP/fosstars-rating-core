```python
import commons


# this is a list of features which are used in the Project Popularity Score
features = [
    'Number of stars for a GitHub repository',
    'Number of watchers for a GitHub repository'
]

class TestVector(commons.BaseTestVector):
        
    def __init__(self, alias = ''):
        super().__init__(features, alias)
            
    def number_of_stars(self, n):
        return self.set('Number of stars for a GitHub repository', n)
    
    def number_of_watchers(self, n):
        return self.set('Number of watchers for a GitHub repository', n)


test_vector_list = commons.TestVectorList(features)
```

```python
all_unknown_values = TestVector().score_from(0.0).score_to(0.1)
test_vector_list.register(all_unknown_values)

test_vector_list.register(
    TestVector()
        .number_of_stars(0)
        .number_of_watchers(0)
        .score_from(0.0)
        .score_to(0.5)
)

test_vector_list.register(
    TestVector()
        .number_of_stars(10)
        .number_of_watchers(0)
        .score_from(0.0)
        .score_to(1.0)
)

test_vector_list.register(
    TestVector()
        .number_of_stars(100)
        .number_of_watchers(0)
        .score_from(0.0)
        .score_to(2.0)
)

test_vector_list.register(
    TestVector()
        .number_of_stars(1000)
        .number_of_watchers(0)
        .score_from(1.0)
        .score_to(2.0)
)

test_vector_list.register(
    TestVector()
        .number_of_stars(10)
        .number_of_watchers(2)
        .score_from(0.0)
        .score_to(1.0)
)

test_vector_list.register(
    TestVector()
        .number_of_stars(100)
        .number_of_watchers(10)
        .score_from(0.0)
        .score_to(2.0)
)

test_vector_list.register(
    TestVector()
        .number_of_stars(1000)
        .number_of_watchers(20)
        .score_from(1.0)
        .score_to(2.0)
)

test_vector_list.register(
    TestVector()
        .number_of_stars(5000)
        .number_of_watchers(20)
        .score_from(4.0)
        .score_to(6.0)
)

test_vector_list.register(
    TestVector()
        .number_of_stars(10000)
        .number_of_watchers(100)
        .score_from(8.0)
        .score_to(10.0)
)

test_vector_list.register(
    TestVector()
        .number_of_stars(20000)
        .number_of_watchers(300)
        .score_from(9.0)
        .score_to(10.0)
)
```

```python
test_vector_list.check()
```

```python
# define test vectors
test_vectors = test_vector_list.make_data_frame()

# store the test vectors to a CSV file
filename = '../../../resources/com/sap/sgs/phosphor/fosstars/model/score/oss/ProjectPopularityScoreTestVectors.csv'
test_vectors.to_csv(filename)

# print out the test vectors
test_vectors
```
